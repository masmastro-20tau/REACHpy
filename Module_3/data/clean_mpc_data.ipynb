```python
import pandas as pd
```

Notebook to update data for NEAs and PHAs from Minor Planet Center. Raw txt data is in fixed width format. Retrieved from [data page](https://www.minorplanetcenter.net/data).

```python
rawfile_path = 'NEA.txt'
colspecs = [(0, 7), (8, 13), (14, 19), (20, 25), (26, 35), (37, 46),
            (48, 57), (59, 68), (70, 79), (80, 91), (92, 103),
            (107, 116), (117, 122), (123, 126), (166, 194), (194, 202)]

colnames = ['Designation', 'Abs Mag', 'Slope Param',
            'Epoch', 'Mean anom (deg)', 'Argument of Perihelion (deg)',
            'Longitude of Ascending Node (deg)', 'Inclination to Ecliptic (deg)',
            'Eccentricity', 'Mean daily motion (deg/day)', 'Semimajor Axis (AU)',
            'Reference', 'N observation', 'N oppositions','Readable Designation',
            'Date of last observation']
            #'Longitude']

raw_data = pd.read_fwf(rawfile_path,colspecs=colspecs,names = colnames, header=None)
raw_data['Date of last observation'] = pd.to_datetime(raw_data['Date of last observation'], format='%Y%m%d')

newfile_path = 'NEA_cleaned.txt'
raw_data.to_csv(newfile_path)
```

```python
newfile_path = 'NEA_cleaned.txt'
cleaned_data = pd.read_csv(newfile_path, parse_dates=['Date of last observation'],header=0,index_col=0)
```
